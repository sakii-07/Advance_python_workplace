```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6])
y = np.array([5,6,7,8,9,10])

plt.plot(x,y,"p-.r",mfc='blue',mec='green',ms=10,lw=2)
dict1={'color':"red",'fontsize':13}
plt.title("student data", loc="center",fontdict=dict1)
plt.xlabel("Id")
plt.ylabel("marks")
plt.grid()
plt.show()
```

```python
total = [10,20,20,40]
plt.pie(x = total, labels=['Python','Java','SQL','HTML'],colors=['hotpink','yellow','orange','skyblue'],
        explode=[0.2,0,0.2,0], startangle=30,counterclock="False")
plt.legend(title="Courses")
plt.show()
```

```python
x = np.array([23,34,42,50,12,28,45,38,19,33])
y = np.random.randint(0,300,10)

plt.plot(x,y,'h-.b',mfc='hotpink', mec='green',lw=1) # ls='dotted'
plt.title("student data",color='red')
plt.grid()
plt.show()
```

```python
total = [30,35,15,20]
plt.pie(total,colors=['skyblue','orange','purple','hotpink'],labels=['Hindi','Marathi','Maths','Science'],
        explode=[0,0.2,0,0.2])
plt.legend(title="Subjects")
plt.show()
```
